### Exercise 1: 

Write a function called writeNLetters() that creates a file named “words.txt” and writes the English alphabet by a specified number of letters on each line. For example, when writeNLetters() is called with an argument value of 3 (i.e., writeNLetters(3)), it should create the following content in “words.txt” file.


ABC
DEF
GHI
JKL
MNO
PQR
STU
VWX
YZ

```python
# Your answer for Exercise 1
def writeNLetters(n: int) -> None:
    abc = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    content = """"""
    iter_ = iter(abc)
    try:
        for _ in range((len(abc) // n + 1)):
            for _ in range(n):
                content += next(iter_)
            content += '\n'
    except StopIteration:
        pass

    with open('words.txt', 'w') as file:
        file.write(content)

writeNLetters(4)
```

### Exercise 2: 
You will write a program that calculates students' GPAs and writes them to a file. You should use suitable collection objects to earn points.
-	The students' grades are stored in the student_grades.csv file. Each line contains one student's grades separated by semicolons. See below for the contents of that csv file (you can also double click on that csv file to see its contents).

The grading system (lettergrade: minimumgrade) and credits for each class are given below. 

•	The grading system:  A : 4.00,  A-: 3.70, B+: 3.30,  B: 3.00, B-: 2.70, C+: 2.30, C: 2.00, C-: 1.70, D+: 1.30, D: 1.00, F 0.00

•	The credits for each class: CS101: 6, ENG101: 4, MATH103: 6, PHYS101: 8, TLL101: 4
  
-	After you calculate the GPAs, write them to another file called  student_gpas.csv together with student ID’s. Your file should look like the following:

Hint:

-	Before you start, check your current working directory with os.getcwd() to decide where to store the files to read. 
-	Read the first line of the student_grades.csv file separately, as it contains the header.
-	Write a function calculate_GPA(student_grades) which takes the information of one student as a parameter, and calculates the GPA and then returns that GPA value. 
-	Do not forget that each line has an end-of-line character at the end and you need to deal with that end-of-line character in your program.


```python
# Your answer for Exercise 2
import csv


grades: dict[str, float] = {'A': 4.00, 'A-': 3.70, 'B+': 3.30, 'B': 3.00, 'B-': 2.70, 'C+': 2.30, 'C': 2.00, 'C-': 1.70,
                            'D+': 1.30, 'D': 1.00, 'F': 0.00}
credit: dict[str, int] = {'CS101': 6, 'ENG101': 4, 'MATH103': 6, 'PHYS101': 8, 'TLL101': 4}


def calculate_GPA(grade: list[float], courses: list[str]) -> float:
    gpa: float = 0.0
    courses: list[int] = [credit[c] for c in courses]
    for x, y in zip(grade, courses):
        gpa += x * y
    gpa /= sum(courses)
    return round(gpa, 2)




with open('student_grades.csv', 'r') as file:
    data: str = csv.reader(file)
    GPAs: str = """"""
    for row in data:
        r: list[str] = row[0].split(';')
        student, grade_list = r[0], r[1:]
        try:
            grade_list = [grades[x] for x in grade_list]
        except KeyError:
            courses = grade_list
        else:
            GPAs += f'{student};{calculate_GPA(grade_list, courses)}\n'

with open('student_gpas.csv', 'w') as file1:
    file1.write(GPAs)
```

```python
# Your answer for Exercise 3

import csv


def add_contact(file_name):
    contact = input("Enter \'s name:").lower()
    email = input("Enter contact\'s email address:")
    phone = input("Enter contact\'s phone number:")
    line = f'{contact};{email};{phone}\n'
    file = open(file_name, 'a')
    file.write(line)
    file.close()


def update_contact_email(file_name):
    contact = input("Enter contact's name: ").lower()
    updated_rows = []
    with open(file_name, mode='r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            if contact in row:
                new_email = input("Enter new email address: ")
                row[1] = new_email
            updated_rows.append(row)

    with open(file_name, mode='w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerows(updated_rows)


def display_contact_details(file_name):
    contact = input("Enter contact's name: ").lower()
    with open(file_name, mode='r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            if contact in row:
                print(f"name:{row[0]}, mail:{row[1]}, phone:{row[2]}")


def main():
    file_name = "address_book.csv"

    while True:
        print("\nAddress Book Management System")
        print("1. Add a new contact")
        print("2. Update a contact's email")
        print("3. Display a contact's details")
        print("4. Exit")
        choice = input("Enter your choice: ")

        match choice:
            case '1':
                add_contact(file_name)
            case '2':
                update_contact_email(file_name)
            case '3':
                display_contact_details(file_name)
            case '4':
                print("Exiting program.")
                break
            case _:
                print("Invalid choice. Please enter a valid option.")


if __name__ == "__main__":
    main()
```
